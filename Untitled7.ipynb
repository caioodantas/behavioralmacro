```julia
'Define valores das constantes'
mm = 1
pstar = 0
eprational=0
epextrapol=0
a1 = 0.5
a2 = -0.2
b1 = 0.5
b2 = 0.05
c1 = 1.5
c2 = 0.5
c3 = 0.5
A = [1 -b2;-a2*c1 1-a2*c2]
B = [b1 0;-a2 a1]
C = [1-b1 0;0 1-a1]
T = 500
TI = 250
K = 50
sigma1 = 0.5
sigma2 = 0.5
sigma3 = 0.5
rho=0.5
rhoout=0.1
rhoinf=0.1
rhotayl=0.1
rhoBH=0.1
epfs=pstar

'Inicializa as séries temporais com 0 em todos os períodos'
p = zeros(T,1)
y = zeros(T,1)
plagt = zeros(T,1)
ylagt = zeros(T,1)
r = zeros(T,1)
epf = zeros(T,1)
epc = zeros(T,1)
ep = zeros(T,1)
ey = zeros(T,1)
CRp = zeros(T,1)
FRp = zeros(T,1)
alfapt = zeros(T,1)
eyfunt = zeros(T,1)
CRy = zeros(T,1)
FRy = zeros(T,1)
alfayt = zeros(T,1)
anspirits = zeros(T,1)
epsilont = zeros(T,1)
etat = zeros(T,1)
ut = zeros(T,1)

alfap=0.5
alfay=0.5
K1=K+1

'Gera variáveis que receberão o impulso'
epsilont2 = zeros(T,1)
etat2 = zeros(T,1)
ut2 = zeros(T,1)
y2 = zeros(T,1)
p2 = zeros(T,1)
r2 = zeros(T,1)
plagt2 = zeros(T,1)
ylagt2 = zeros(T,1)
epsilont2[1] = sigma1*randn() + sigma1
etat2[1]= sigma2*randn() + sigma2
ut2[1] =sigma3*randn() +sigma3



for t in 2:500

  epsilont[t] = rhoout*epsilont[t-1] + sigma1*randn()
  etat[t]= rhoinf*etat[t-1] + sigma2*randn()
  ut[t] = rhotayl*ut[t-1] + sigma3*randn()
    
  epsilont2[t] = rhoout*epsilont2[t-1] + sigma1*randn()
  etat2[t]= rhoinf*etat2[t-1] + sigma2*randn()
  ut2[t] = rhotayl*ut2[t-1] + sigma3*randn()  
  shocks2 = [eta2;a2*u2+epsilont2[t]]

  epsilon = epsilont[t]
  eta = etat[t]
  u = ut[t]
  shocks = [eta;a2*u+epsilon]
  epcs = p[t-1]
  if eprational==1
    epcs=pstar
  end
  eps=alfap*epcs+(1-alfap)*epfs
  if epextrapol==1
    eps=p(t-1)
  end
  eychar=y[t-1]
  eyfun=0+randn()/2
  eyfunt[t]=eyfun
  eys=alfay*eychar+(1-alfay)*eyfun
  forecast = [eps[];eys]
  plag=p[t-1]
  ylag=y[t-1]
  rlag=r[t-1]
  lag = [plag;ylag]
  smooth = [0;a2*c3]
  D = B*forecast + C*lag + smooth*rlag + shocks
  D2= B*forecast + C*lag + smooth*rlag + shocks2
  X = A\D
  X2=A\D2
  p[t]= X[1,1]
  y[t]= X[2,1]
  r[t]= c1*p[t]+c2*y[t]+c3*r[t-1]+u
  p2[t]= X[1,1]
  y2[t]= X[2,1]
  r2[t]= c1*p2[t]+c2*y2[t]+c3*r2[t-1]+u
  plagt[t]=p[t-1]
  ylagt[t]=y[t-1]
  plagt2[t]=p[t-1]
  ylagt2[t]=y[t-1]
  CRp[t] = rho*CRp[t-1] - (1-rho)*(epcs-p[t])^2
  FRp[t] = rho*FRp[t-1] - (1-rho)*(epfs-p[t])^2
  CRy[t] = rho*CRy[t-1] - (1-rho)*(eychar-y[t])^2
  FRy[t] = rho*FRy[t-1] - (1-rho)*(eyfun-y[t])^2
  alfap = rhoBH*alfapt[t-1]+(1-rhoBH)*exp(mm*CRp[t])/(exp(mm * CRp[t]) + exp(mm * FRp[t]))
  alfay = rhoBH*alfayt[t-1]+(1-rhoBH)*exp(mm*CRy[t])/(exp(mm * CRy[t]) + exp(mm * FRy[t]))
  alfapt[t] = alfap
  alfayt[t] = alfay
  if eychar>0
    anspirits[t]=alfay
  end
  if eychar<0
    anspirits[t]=1-alfay
  end
end
```
